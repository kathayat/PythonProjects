```python
 #Import modules 
    
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
#Scrape data
#Variable for base url

base_url = 'https://understat.com/match/'
match = str(input('Please enter match id:'))

url = base_url + match
```

```python
url
```

```python
res = requests.get(url)

soup = BeautifulSoup(res.content,'lxml')
scripts = soup.find_all('script')
```

```python
scripts
```

```python
#Get only shots data

strings = scripts[1].string
```

```python
strings
```

```python
#Strip symbols

index_start = strings.index("('")+2
index_end = strings.index("')")

json_data = strings[index_start:index_end]
json_data = json_data.encode('utf8').decode('unicode_escape')

#Convert string to json

data = json.loads(json_data)
```

```python
data
```

```python
x = []
y = []
xg = []
team = []

data_away = data['a']
data_home = data['h']

#Looping in the data frame 
for index in range(len(data_home)):
    for key in data_home[index]:
        if key == 'X':
            x.append(data_home[index][key])
        if key == 'Y':
            y.append(data_home[index][key])
        if key == 'xG':
            xg.append(data_home[index][key])
        if key == 'h_team':
            team.append(data_home[index][key])
            
            
for index in range(len(data_away)):
    for key in data_away[index]:
        if key == 'X':
            x.append(data_away[index][key])
        if key == 'Y':
            y.append(data_away[index][key])
        if key == 'xG':
            xg.append(data_away[index][key])
        if key == 'a_team':
            team.append(data_away[index][key])
```

```python
#Create the dataframe

col_names = ['x','y','xg','team']
df = pd.DataFrame([x,y,xg,team],index = col_names)
```
