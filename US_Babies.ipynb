```python
#Top female and male names in california
ca_babies = state_babies.loc[state_babies['State']=='CA',:]
```

```python
ca_babies.head(5)
```

```python
def popular(s):
    return s.iloc[0]

# Receives s, a panda series containing baby names in order of highest count to lowest. Returns most popular names in desc.
```

```python
#Groupby year and gender
ca_grouped = ca_babies.sort_values('Count', ascending= False).groupby(['Year','Gender']).agg(popular)
#Returns groupby object, not data frame
```
